```python
import pandas as pd
```

```python
consumers = pd.read_csv('Dataset/consumers.csv')
consumers
```

```python
consumers.isna().sum()
```

```python
consumers.info()
```

```python
#Filling null values with imputation
for col in ['Smoker','Transportation_Method','Marital_Status','Children','Occupation','Budget']:
    consumers[col].fillna(consumers[col].mode()[0], inplace=True)

consumers.isna().sum()
```
